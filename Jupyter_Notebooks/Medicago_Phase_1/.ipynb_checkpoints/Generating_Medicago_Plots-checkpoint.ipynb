#Generating QPSI Timeseries figures#

```python
#Import the library containing the functions we will be using
#Make sure this file is in your current directory
import BioinformaticsLibrary as BL #This can be found in the Python Scripts directory. Make sure the BioinformaticsLibrary.py script is in your working directory
import matplotlib.pyplot as plt

#Obtain Medicago DEG counts
med_DEG_count_DF = BL.count_DEGs(DEG_file='Compiled_DEGs_wAnnotation.xlsx', 
                   LFC_suffix='_log2FC', padj_suffix='_Padj', 
                   L2FC_threshold=1, adjP_threshold=0.05)

```

```python
#Separate Hg from Cd treated samples
hg_count = med_DEG_count_DF[med_DEG_count_DF['Comparison'].str.contains('Hg')]
cd_count = med_DEG_count_DF[med_DEG_count_DF['Comparison'].str.contains('Cd')]
```

```python
#Plot DEGs on 2 separarte plots, Hg and Cd
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15, 8), gridspec_kw={'width_ratios': [5, 6]})
fig.suptitle('Total Differentially Expressed Genes by Treatment')
hg_count.set_index('Comparison').plot(ax=ax1, kind='bar', stacked=True, color=['red', 'steelblue'], width=0.8)
cd_count.set_index('Comparison').plot(ax=ax2, kind='bar', stacked=True, color=['red', 'steelblue'], width=0.8)

# Add Title and Labels
plt.xlabel('Treatment')
plt.ylabel('Total DEGs')
plt.tick_params(axis='both', which='major', labelsize=8)
plt.savefig('Medicago_stacked_split_DEG_by_total.pdf', bbox_inches='tight')

#Export DEG summary table
med_DEG_count_DF.to_csv('Medicago_v5_DEG_summary_table.csv', index=False)
```
